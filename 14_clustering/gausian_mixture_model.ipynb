```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans

clusters_std = [2.5, 0.5]

# Create a synthetic dataset with two well-separated clusters
X, y = make_blobs(n_samples=300, centers=2, random_state=2, cluster_std=clusters_std)

# Standardize the dataset
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# Visualize the dataset
plt.scatter(X_std[:, 0], X_std[:, 1], c=y, cmap='viridis', s=40, edgecolors='k', linewidth=0.5)
plt.title("Synthetic 2D Dataset with Two Clusters")
plt.xlabel("Feature 1 (Standardized)")
plt.ylabel("Feature 2 (Standardized)")
plt.show()

# Fit GMM
gmm = GaussianMixture(n_components=2, random_state=42)
gmm_labels = gmm.fit_predict(X_std)

# Visualize GMM clusters
plt.scatter(X_std[:, 0], X_std[:, 1], c=gmm_labels, cmap='viridis', s=40, edgecolors='k', linewidth=0.5)
plt.title("GMM Clustering")
plt.xlabel("Feature 1 (Standardized)")
plt.ylabel("Feature 2 (Standardized)")
plt.show()

# Fit K-Means
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X_std)

# Visualize K-Means clusters
plt.scatter(X_std[:, 0], X_std[:, 1], c=kmeans_labels, cmap='viridis', s=40, edgecolors='k', linewidth=0.5)
plt.title("K-Means Clustering")
plt.xlabel("Feature 1 (Standardized)")
plt.ylabel("Feature 2 (Standardized)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
```

```python
# Set random seed for reproducibility
np.random.seed(21)
```

```python
# Generate synthetic data
n_samples = 1500
```

```python
# Define parameters for three Gaussian clusters
mean1 = [-2, 2]
cov1 = [[1, 0.9], [0.9, 1]]  # High correlation
mean2 = [5, 5]
cov2 = [[1, 0], [0, 1]]  # No correlation
mean3 = [-5, 5]
cov3 = [[1, -0.8], [-0.8, 1]]  # Negative correlation
```

```python
# Generate points for each cluster
X1 = np.random.multivariate_normal(mean1, cov1, n_samples // 3)
X2 = np.random.multivariate_normal(mean2, cov2, n_samples // 3)
X3 = np.random.multivariate_normal(mean3, cov3, n_samples // 3)

# Combine the points to create the dataset
X = np.vstack((X1, X2, X3))
```

```python
# Plot the generated data
plt.scatter(X[:, 0], X[:, 1], s=10)
plt.title("Generated Data")
plt.show()
```

```python
# Fit a K-means model
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X)
```

```python
# Plot GMM clustering results
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, s=10, cmap='viridis')
plt.title("KMeans Clustering")
plt.show()
```

```python
gmm = GaussianMixture(n_components=3, random_state=42)
gmm_labels = gmm.fit_predict(X)
```

```python
# Plot GMM clustering results
plt.scatter(X[:, 0], X[:, 1], c=gmm_labels, s=10, cmap='viridis')
plt.title("GMM Clustering")
plt.show()
```
