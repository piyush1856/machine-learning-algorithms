```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans

clusters_std = [2.5, 0.5]

# Create a synthetic dataset with two well-separated clusters
X, y = make_blobs(n_samples=300, centers=2, random_state=2, cluster_std=clusters_std)

# Standardize the dataset
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# Visualize the dataset
plt.scatter(X_std[:, 0], X_std[:, 1], c=y, cmap='viridis', s=40, edgecolors='k', linewidth=0.5)
plt.title("Synthetic 2D Dataset with Two Clusters")
plt.xlabel("Feature 1 (Standardized)")
plt.ylabel("Feature 2 (Standardized)")
plt.show()

# Fit GMM
gmm = GaussianMixture(n_components=2, random_state=42)
gmm_labels = gmm.fit_predict(X_std)

# Visualize GMM clusters
plt.scatter(X_std[:, 0], X_std[:, 1], c=gmm_labels, cmap='viridis', s=40, edgecolors='k', linewidth=0.5)
plt.title("GMM Clustering")
plt.xlabel("Feature 1 (Standardized)")
plt.ylabel("Feature 2 (Standardized)")
plt.show()

# Fit K-Means
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X_std)

# Visualize K-Means clusters
plt.scatter(X_std[:, 0], X_std[:, 1], c=kmeans_labels, cmap='viridis', s=40, edgecolors='k', linewidth=0.5)
plt.title("K-Means Clustering")
plt.xlabel("Feature 1 (Standardized)")
plt.ylabel("Feature 2 (Standardized)")
plt.show()
```
